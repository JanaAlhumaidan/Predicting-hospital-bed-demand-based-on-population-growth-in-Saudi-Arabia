```python
import pandas as pd

df = pd.read_csv("population_per_region.csv", encoding="utf-16")
df.head()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

future_predictions = []

regions = df["Region"].unique()

for region in regions:
    region_df = df[df["Region"] == region]

    X = region_df[["Year"]]
    y = region_df["Population"]

    model = LinearRegression()
    model.fit(X, y)

    future_years = pd.DataFrame({
        "Year":[2023,2024,2025,2026]
    })

    preds = model.predict(future_years)

    temp = future_years.copy()
    temp["Region"] = region
    temp["Predicted_Population"] = preds

    future_predictions.append(temp)

future_population = pd.concat(future_predictions)
print(future_population)
```
